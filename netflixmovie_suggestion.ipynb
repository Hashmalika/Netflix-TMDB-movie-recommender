feature  selection part

```python
movie.columns
```

```python
movies=movie[['id','title','overview','genres']]
movies
```

```python
movies['tags']=movies['overview']+movies['genres']
movies
```

```python
new_data=movies.drop(columns=['overview','genres'])
new_data=new_data.head(10000)
new_data
```

1)Bag of Word
2)TFIDF  are  used  to  convert  tags  into  vectors

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(max_features=2966,stop_words='english')
```

```python
vector=cv.fit_transform(new_data['tags'].values.astype('U')).toarray()
```

```python
vector
```

```python
vector.shape
```

here  we  use  cosine  similarity  (eg. 10000
 x axis now  distance  between  2 movie  on 2 axis  if  angle is low  then  cosine  similairty)

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vector)
```

```python
similarity
```

```python
new_data[new_data['title']=='Interstellar'].index[0]
```

```python
new_data
```

```python
distance=sorted(list(enumerate(similarity[2])),reverse=True,key=lambda vector:vector[1])
```

```python
for i in distance[1:6]:
  print(new_data.iloc[i[0]].title)
```

```python
def recommend(movies):
  index=new_data[new_data['title']==movies].index[0]
  distance=sorted(list(enumerate(similarity[index])),reverse=True,key=lambda vector:vector[1])
  for i in distance[1:6]:
    print(new_data.iloc[i[0]].title)
```

```python
recommend("Rope")
```
